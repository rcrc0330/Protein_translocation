```python
a = math.acos((((2*(dh**2))-((ds**2)))/(2*(dh**2))))


Z = []
t=-w/2
while(t<=w/2):
    Z.append(t)
    t=t+ds



angle=0
ans=([])
temp=[]
while(angle<2*math.pi):
    x=(dh)*(math.cos(angle))
    y=(dh)*(math.sin(angle))
    for z in Z:
        print(z)
        ans.append(([1,x,y,z]))
    angle=angle+a





X = []
Y = []
tx = -l/2
ty = -d/2
while(tx<=l/2):
    X.append(tx)
    tx=tx+db
while(ty<=d/2):
    Y.append(ty)
    ty=ty+db

for x in X:
    for y in Y:
        if((x**2)+(y**2)>=(db/2 + dh)**2):
            ans.append([2,x,y,-w/2])
            ans.append([2,x,y,w/2])



n = len(ans)
```
