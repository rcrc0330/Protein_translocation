```python
import numpy as np
import math
l = float(input("Length of the surface:"))
d = float(input("Depth of the surface:"))
w = float(input("Width of the hole:"))
dh = float(input("Diameter of the hole:"))
db = float(input("Diameter of the bigger spheres:"))
ds = float(input("Diameter of the smaller spheres:"))
```

```python
 a = math.acos((((2*(dh**2))-((ds**2)))/(2*(dh**2))))
```

```python
print(a)
```

```python
Z = []
t=-w/2
while(t<=w/2):
    Z.append(t)
    t=t+ds
```

```python
print(Z)
```

```python
angle=0
ans=([])
temp=[]
while(angle+a<2*math.pi):
    x=(dh/2)*(math.cos(angle))
    y=(dh/2)*(math.sin(angle))
    temp.append(np.array([x,y]))
    for z in Z:
        print(z)
        ans.append(([1,x,y,z]))
    angle=angle+a
```

```python
print(ans)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dis = ds/2 + db/2
X = []
Y = []
tx = -l/2
ty = -d/2
while(tx<=l/2):
    X.append(tx)
    tx=tx+db
while(ty<=d/2):
    Y.append(ty)
    ty=ty+db
```

```python
for x in X:
    for y in Y:
        if((x**2)+(y**2)>(dh/2+dis)**2):
            ans.append([1,x,y,-w/2])
            ans.append([1,x,y,w/2])
```

```python
print(ans)
```

```python
s = np.array(ans)
```
